```python
# Tareaita 2: Grafica funcion de interes

# Importo las bibliotecas que me van a permitir crear la grafica
import numpy as np
import matplotlib.pyplot as plt

# Defino las variables C e i
C=50000  # Capital
i=0.05  # Tasa de interes (5%)

# Defino el tiempo (temporalidad de pagos "periodos")
t= np.linspace(0, 12, 1000)  # Tomo 1000 puntos entre 0 y 12 periodos

# Defino las formulas de interes simple y compuesto dadas en la clase
M_simple=C*(1+i*t)
M_compuesto=C*(1+i)**t

# Creo la figura y los ejes de la grafica
plt.figure(figsize=(10, 6))
plt.style.use('seaborn-darkgrid')

# Grafico las curvas
plt.plot(t, M_simple, label="Interés Simple", color='blue')
plt.plot(t, M_compuesto, label="Interés Compuesto", color='red')

# Añado los títulos y etiquetas para la grafica
plt.title('Interés Simple vs Interés Compuesto')
plt.xlabel('Periodos (t)')
plt.ylabel('Monto Final (M)')
plt.legend()

# Mostrar la gráfica
plt.show()

# Ya podemos observar la comparación (:
```
